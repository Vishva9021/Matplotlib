```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories.

np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category': np.random.choice(categories, size=100),
    'Value': np.random.randn(100) * 10 + 50  # Random values with some noise
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Create a box plot to visualize the distribution of values across different categories
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Category'] == cat]['Value'] for cat in categories], 
            labels=categories, 
            patch_artist=True)

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Numerical Values Across Different Categories')

# Show the plot
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
#heatmap.

np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 100
x1 = np.random.randn(n_samples)
x2 = 0.8 * x1 + np.random.randn(n_samples) * 0.2  # x2 is correlated with x1
x3 = 0.5 * x1 + 0.5 * x2 + np.random.randn(n_samples) * 0.5  # x3 is correlated with x1 and x2

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3
})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)

# Add labels and a title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function 


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add labels and a title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line to the figure
p.line(x, y, line_width=2, color='blue', legend_label='Sine Wave')

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.


# Generate random data
np.random.seed(42)
n = 100
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(5, 50, n)  # Marker sizes
colors = np.random.rand(n)  # Random colors

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add a scatter plot to the figure
p.scatter(
    x='x',
    y='y',
    size='sizes',
    color='colors',
    source=df,
    legend_label='Data Points',
    fill_alpha=0.6
)

# Show the plot in the notebook
output_notebook()
show(p)

```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

# Prepare the dataset
data = {
    'Fruit': ['Apple', 'Banana','banana' 'Date', 'pears'],
    'Count': [20, 25, 30, 35]
}
df = pd.DataFrame(data)

# Create a Bokeh figure
p = figure(x_range=df['Fruit'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', 
           plot_height=350, plot_width=600, toolbar_location=None, tools='')

# Add a vertical bar renderer
p.vbar(x='Fruit', top='Count', width=0.9, source=df, color='green', legend_label='Count')

# Add labels and customize appearance
p.xgrid.grid_line_color = None
p.yaxis.axis_label = 'Count'
p.xaxis.axis_label = 'Fruit'
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.orientation = 'horizontal'
p.legend.location = 'top_center'

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Data Distribution", x_axis_label='Value', y_axis_label='Frequency',
           plot_height=400, plot_width=600, toolbar_location=None)

# Add a quad renderer to the figure
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white")

# Customize appearance
p.grid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.




# Generate random data for heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays and create a DataFrame
df = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y',
           plot_height=400, plot_width=400, toolbar_location=None)

# Define a color mapper
mapper = linear_cmap(field_name='value', palette='Viridis256', low=min(df['value']), high=max(df['value']))

# Add a rect renderer to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=df, fill_color=mapper, line_color=None)

# Customize appearance
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.major_label_orientation = 'vertical'
p.yaxis.major_label_orientation = 'horizontal'
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Show the plot in the notebook
output_notebook()
show(p)
```
